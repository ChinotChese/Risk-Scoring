# Volatility (Average True Range)

```python
data = yf.download('NFLX', period="12mo", interval="1d")
```

```python
data.head()
```

**Formula**
$$TR = max[|H-L|, |H-C_p|, |L-C_p|]$$
$$ATR = \frac{1}{n}\Sigma_{i=1}^{n}TR_i$$
*$TR_i$: particular true range\
*H: high price today\
*L: low price today\
*$C_p$: close price yesterday

```python
high_low = data['High'] - data['Low']
```

```python
high_cp = abs(data['High'] - data['Close'].shift())
```

```python
low_cp = abs(data['Low'] - data['Close'].shift())
```

```python
df = pd.concat([high_low, high_cp, low_cp], axis=1)
df.head()
```

```python
TR = np.max(df, axis=1)
TR.head()
```

```python
ATR = TR.rolling(14).mean()
ATR
```

```python
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
fig, ax = plt.subplots()
ATR.plot(ax=ax)
ax2 = data['Close'].plot(ax=ax, secondary_y=True, alpha=.3)
ax.set_ylabel("ATR")
ax2.set_ylabel("Price")
fig.savefig('./atr_nflx')
```
